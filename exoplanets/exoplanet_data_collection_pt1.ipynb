# NASA's archive on Exo-planets.  Part 1 - Data Collection

```python
#start by getting the data from NASA's exoplanet-archive

import pandas as pd
import json
import requests
```

```python
#url do get data

url ="https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+*+from+pscomppars&format=json"
```

```python
r = requests.get(url)
```

```python
#get data in json format
r_json=  r.json()
```

```python
#slice data to see what is looks like.  As you can see it's pretty big.
r_json[0]
```

```python
#convert the data from dictionary-style to dataframe.

col_headings = [c for c in r_json[0]]

comp_dict ={}

for d in col_headings:
    comp_dict[d] = [r_json[i][d] for i in range(len(r_json))]
```

```python
df = pd.DataFrame(comp_dict)
df
```

```python
#would be good to update columns names to something with more meanining.  Luckily NASA provided a column mapping table that translates field names into something more understandable

column_mapping = pd.read_csv(r'C:\Exoplanet_Archive_Column_Mapping_CSV.csv')
column_mapping
```

```python
#the dataset we got relates to Planetary Systems Composite Parameters (PSCP).  Therefore we will only need columns 3 and 4.  We don't need th rest of thecolumn
```

```python
new_col_mapping = column_mapping.iloc[1:, 2:4]
```

```python
#updating the 2nd column name
new_col_mapping = new_col_mapping.rename(columns={'Unnamed: 3':'Column Headers'
})

new_col_mapping
```

```python
#something not visible, but there is a blank space at the start of each field item  (e.g ' pl_name' and not 'pl_name')

new_col_mapping.iloc[0][0]
```

```python
#to correct, convert to string then apply lambda

new_col_mapping['Planetary Systems Composite Parameters (PSCP)'] = new_col_mapping['Planetary Systems Composite Parameters (PSCP)'].astype(str).apply(lambda x:x.replace(' ',''))
```

```python
#check to see if fixed
new_col_mapping.iloc[0][0]
```

```python
#lets lookup the columns headings in our dataframe with the new mapping table
df_cols = pd.merge(pd.DataFrame(df.columns, columns={'col title'}), new_col_mapping, how='left', left_on='col title', right_on='Planetary Systems Composite Parameters (PSCP)')
df_cols
```

```python
#i see some NaN in the data.  For now, we can exlude these as my column headings from JSON data isn't in the column mapping

df_cols.dropna()
```

```python
x, y = df_cols.dropna().iloc[:,1],df_cols.dropna().iloc[:,2]
```

```python
#lets use a zip to get into dictionary fomart, which we can then apply to rename all column in data frame

x, y = df_cols.dropna().iloc[:,1],df_cols.dropna().iloc[:,2]

new_headers = dict(zip(x,y))
new_headers
```

```python
df2=df[x].rename(columns= new_headers)
```
