# NASA's archive on Exo-planets. Part 2 - Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#lets loads our dataset

df = pd.read_excel(r'C:\nasa_exoplanet_data.xlsx')
```

```python
#get column headings and see what we can analyse

list(df.columns)

```

```python
columns = ['Planet Name',
'Discovery Year',
'Orbital Period [days]',
'Equilibrium Temperature [K]',
'Planet Density [g/cm**3]',
'Planet Mass or Mass*sin(i) [Earth Mass]',
'Distance [pc]',
'Stellar Rotational Period [days]',
'Radial Velocity Amplitude [m/s]',
'Planet Radius [Earth Radius]',
'Stellar Effective Temperature [K]',
'Stellar Radius [Solar Radius]',
'Stellar Mass [Solar mass]',
'Stellar Density [g/cm**3]',
'Number of Stars',
'Number of Planets',
'Number of Moons']

df2 = df[columns]

df2
```

```python
#lets see how many planets were discovered over time

ax = sns.countplot(x=df2['Discovery Year'], color='Blue', palette='Blues')
sns.set(rc={'figure.figsize':(20,8)})
# plt.text(2016, 1517,"test", fontsize=12)

ax.text(24, 1500, "Significant increase in new planets discovered in 2016 due to the launch of the Kepler Satelittie", 
       fontsize = 14,          # Size
       fontstyle = "oblique",  # Style
       color = "black",          # Color
       ha = "left", # Horizontal alignment
       va = "bottom") 
```

```python
#plotting planets density mass from the dataset

sns.scatterplot(data=df2, x='Planet Mass or Mass*sin(i) [Earth Mass]', y='Planet Density [g/cm**3]')
```

```python
#lets focus on where the bulk of the data is which is between 0 and less than 50.  Interestingly, there seems to be a striking line occuring.
sns.set_theme(style="whitegrid")
sns.scatterplot(data=df2, x='Planet Mass or Mass*sin(i) [Earth Mass]', y='Planet Density [g/cm**3]')
plt.xlim(0,50)
plt.ylim(0,10)
plt.title("Planet density vs planet Mass", size=20)
```

```python
#what planets have similar tempatures and size to our sun and simimlar density to earth
temp_sun = 5772
temp_sun_min = 0.8* temp_sun
temp_sun_max = 1.2 * temp_sun
```

```python
query1=df2[
    (df2['Stellar Effective Temperature [K]'] > temp_sun_min) &
    (df2['Stellar Effective Temperature [K]'] < temp_sun_max) &
    (df2['Stellar Mass [Solar mass]'] > 0.8) &
    (df2['Stellar Mass [Solar mass]'] < 1.2) &
    (df2['Planet Density [g/cm**3]'] > 5) &
    (df2['Planet Density [g/cm**3]'] < 6)
    ]
```

```python
cmap = sns.cubehelix_palette(rot=-0.2, as_cmap=True)
```

```python
sns.relplot(
    data = query1,
    x = 'Stellar Effective Temperature [K]',
    y='Planet Density [g/cm**3]', size ="Stellar Mass [Solar mass]",
    palette=cmap, sizes=(4,250))
```

```python
#We can see that smaller stars have lower temperature compared to larger ones.  Distrubtion is clustered around a specific area

sns.set_theme(style="whitegrid")

g=sns.relplot(
    data=df2,
    x="Stellar Radius [Solar Radius]", y="Stellar Effective Temperature [K]",
    hue="Stellar Density [g/cm**3]", size="Stellar Mass [Solar mass]",
    palette=cmap, sizes=(10, 500), height=10)

plt.xlim(0,4)
plt.ylim(2000,10000)


g.despine(left=True, bottom=True)

```
